```python
import matplotlib.pyplot as plt
import pandas as pd

ticker = 2330
df= pd.read_csv('chips/{}.csv'.format(ticker), parse_dates=True, index_col=1)
del df['Unnamed: 0']

#print(df.head())

#graph
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)

ax1.plot(df.index, df['prc_C'])
ax2.plot(df.index, df['ratio_15'])

plt.show()
```
