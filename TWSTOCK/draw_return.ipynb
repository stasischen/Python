```python
from pandas_datareader import data 
import matplotlib.pyplot as plt   
import pandas as pd                
%matplotlib inline

year = 2001
tickers = '1101'


data = pd.ExcelFile('weeks_return/{}R.xlsx'.format(year))
df = data.parse("Sheet1")

filter = df['ticker'] == tickers

stock_info = df[filter]
stock_info['return'] = 1 + (stock_info['return']/100)
stock_info.index = stock_info['date']

#stock_info = pd.concat([stock_info, stock_return], axis=1)
stock_info['return'].cumprod().plot(color='blue')
```
