```python
from pandas_datareader import data 
import matplotlib.pyplot as plt   
import pandas as pd   
import csv
%matplotlib inline

tickers = '1101'

df = pd.read_csv('weeks_return/weekly1.csv', encoding = 'utf8')
    
#證券代碼 簡稱 年月日 週報酬率 %
filter = df['證券代碼'] == tickers

stock_info = df[filter]
stock_info['週報酬率 %'] = 1 + (stock_info['週報酬率 %']/100)
stock_info.index = stock_info['年月日']

#stock_info = pd.concat([stock_info, stock_return], axis=1)
stock_info['週報酬率 %'].cumprod().plot(color='blue')
```
